```python
x = df.drop(['diagnosis'], axis = 1)
y = df['diagnosis']
```

```python
from sklearn.model_selection import train_test_split

x_train_feature, x_test_feature, y_train_target, y_test_target = train_test_split(x,y,test_size = 0.25, random_state = 40, shuffle = True)
```

```python
x_train_feature.shape
```

```python
y_train_target.shape
```

```python
y_test_target.shape
```

```python
x_train, x_val, y_train, y_val = train_test_split(x_train_feature, y_train_target, 
                                                    test_size = .2, 
                                                    random_state=12)
print(x_train.shape)
print(x_val.shape)
print(y_train.shape)
print(y_val.shape)
```

```python
from imblearn.over_sampling import SMOTE

sm = SMOTE(random_state=42, sampling_strategy=0.8)

x_train_res, y_train_res = sm.fit_resample(x_train, y_train)
y_train_res.shape
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
y_train_res_encoded = le.fit_transform(y_train_res)
y_val_encoded = le.transform(y_val)
y_test_target_encoded = le.transform(y_test_target)
```

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
x_train_res_scaled = ss.fit_transform(x_train_res)
x_val_scaled = ss.transform(x_val)
x_test_feature_scaled = ss.transform(x_test_feature)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import recall_score
clf_rf = RandomForestClassifier(n_estimators=25, random_state=12)
clf_rf.fit(x_train_res_scaled, y_train_res_encoded)

print('Validation Results:')
print(clf_rf.score(x_val_scaled, y_val_encoded))
print(recall_score(y_val_encoded, clf_rf.predict(x_val_scaled)))

print('\nTestResults:')
print(clf_rf.score(x_test_feature_scaled, y_test_target_encoded))
print(recall_score(y_test_target_encoded, clf_rf.predict(x_test_feature_scaled))) 
```

```python
from sklearn.ensemble import GradientBoostingClassifier
lr_list = [0.05, 0.075, 0.1, 0.25, 0.5, 0.75, 1]

for learning_rate in lr_list:
    gb = GradientBoostingClassifier(n_estimators=20, learning_rate=learning_rate, max_features=2, max_depth=2, random_state=0)
    gb.fit(x_train_res_scaled, y_train_res_encoded)

    print('Validation Results:')
    print(gb.score(x_test_scaled, y_test_encoded))
    print(recall_score(y_test_encoded, gb.predict(x_test_scaled)))
    print('TestResults:')
    print(gb.score(x_test_feature_scaled, y_test_target_encoded))
    print(recall_score(y_test_target_encoded, gb.predict(x_test_feature_scaled))) 
    print('------------------------------------------')
```

```python
models = {}

# Logistic Regression
from sklearn.linear_model import LogisticRegression
models['Logistic Regression'] = LogisticRegression()

# Support Vector Machines
from sklearn import svm
models['Support Vector Machines'] = svm.SVC(gamma = 'auto')

# Decision Trees
from sklearn.tree import DecisionTreeClassifier
models['Decision Trees'] = DecisionTreeClassifier()

# Random Forest
from sklearn.ensemble import RandomForestClassifier
models['Random Forest'] = RandomForestClassifier()

# Naive Bayes
from sklearn.naive_bayes import GaussianNB
models['Naive Bayes'] = GaussianNB()

from sklearn.metrics import accuracy_score, precision_score, recall_score
from sklearn.metrics import f1_score
accuracy, precision, recall = {}, {}, {}
for key in models.keys():
    
    # Fit the classifier model
    models[key].fit(x_train_res_scaled, y_train_res_encoded)
    
    # Prediction 
    predictions = models[key].predict(x_test_feature_scaled)
    
    # Calculate Accuracy, Precision and Recall Metrics
    accuracy[key] = accuracy_score(predictions, y_test_target_encoded)
    precision[key] = precision_score(predictions,y_test_target_encoded)
    recall[key] = recall_score(predictions, y_test_target_encoded)
```

```python
import pandas as pd

df_model = pd.DataFrame(index=models.keys(), columns=['Accuracy', 'Precision', 'Recall'])
df_model['Accuracy'] = accuracy.values()
df_model['Precision'] = precision.values()
df_model['Recall'] = recall.values()

df_model
```

```python
from sklearn.metrics import confusion_matrix

y_pred= models['Naive Bayes'].predict(x_test_feature_scaled)

cm = confusion_matrix(y_test_target_encoded, y_pred)
cm
```
