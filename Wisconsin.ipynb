```python
x = df.drop('diagnosis', axis = 1)
y = df.diagnosis
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
y.shape
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.25, random_state = 40)
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
y_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
x_train = ss.fit_transform(x_train)
x_test = ss.transform(x_test)
```

```python
#great!!! we see the score is high
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report

le = LogisticRegression()
le.fit(x_train, y_train)
y_pred = le.predict(x_test)
print("Training score:", le.score(x_train,y_train))
print("Training score:", le.score(x_test,y_test))

cr = classification_report(y_test, y_pred)
cr
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics
from sklearn.metrics import confusion_matrix

dtc = DecisionTreeClassifier()
dtc.fit(x_train, y_train)
y_pred = dtc.predict(x_test)

print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```

```python
from sklearn.ensemble import GradientBoostingClassifier
lr_list = [0.05, 0.075, 0.1, 0.25, 0.5, 0.75, 1]

for learning_rate in lr_list:
    gb = GradientBoostingClassifier(n_estimators=20, learning_rate=learning_rate, max_features=5, max_depth=3, random_state=0)
    gb.fit(x_train, y_train)

    print("Learning rate: ", learning_rate)
    print("Accuracy score (training): {0:.4f}".format(gb.score(x_train, y_train)))
    print("Accuracy score (test): {0:.4f}".format(gb.score(x_test, y_test)))
```

```python
from sklearn.neural_network import MLPClassifier

mlp = MLPClassifier(random_state=10, max_iter=400).fit(x_train, y_train)
print(mlp.score(x_train,y_train))
print(mlp.score(x_test,y_test))
```

```python
#Best Model
from sklearn.svm import SVC

sv = SVC(gamma = 'auto', C=5, kernel = 'rbf')
sv.fit(x_train, y_train)
print("svc test score is:", sv.score(x_test, y_test))
```
